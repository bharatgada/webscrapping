```python
from bs4 import BeautifulSoup
import pandas as pd
import requests,openpyxl
from openpyxl import Workbook
```

```python
excel = openpyxl.Workbook()
print(excel.sheetnames)
sheet = excel.active
sheet.title = 'One-Day Highest Run scores'
print(excel.sheetnames)
sheet.append(['Player Name', 'Country', 'Runs Scored', 'Played Year'])
```

   # 1997 ONE-DAY MOST RUNS

```python
source = requests.get('https://stats.espncricinfo.com/ci/engine/records/batting/most_runs_career.html?class=2;id=1997;type=year')
source.raise_for_status()

soup = BeautifulSoup(source.text,'html.parser')

players = soup.find('tbody').find_all('tr')

for player in players:
    name = player.find('td', class_="left").a.text
    country = player.find('td', class_="left").text.split(' ')[-1].strip('()')
    score = player.find_all('td', nowrap="nowrap")[4].b.text
    year = 1997
    print(name, country, score, year)
    sheet.append([name, country, score, year])
```
